```python


# Assuming your files are in the current working directory
file_prefix = "synchronized_data"
file_extension = ".csv"
output_prefix = "combined_data_with_interpolation"
num_files = 100

# Loop through the file names and read, process, and save each file
for i in range(1, num_files + 1):
    # Read the CSV file
    file_name = file_prefix + str(i) + file_extension
    df = pd.read_csv(file_name)

    # Apply your processing logic
    df = df[df["PWM1"] > 900]
    time_values = df['Time']
    new_time_values = np.linspace(time_values.min(), time_values.max(), 2000)
    interpolated_data = pd.DataFrame({'Time': new_time_values})
    for column in df.columns:
        if column != 'Time':
            spline = CubicSpline(time_values, df[column], bc_type='natural')
            interpolated_data[column] = spline(new_time_values)
    interpolated_data.to_csv(output_prefix+str(i)+file_extension, index=False)
```
